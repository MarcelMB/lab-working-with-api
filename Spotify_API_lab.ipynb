```python
import requests
import json
import base64
import pandas as pd
import getpass
```

```python
#just check general connection, but we want to acces specific playlist so we need further authorization
spotify=requests.get("https://api.spotify.com")
spotify.status_code
```

```python
# created those variables in dashboard area of spotify devlopers page
client_id='e1d8bf3111fa46718e6cdb8b109022f0'
client_secret = getpass.getpass()


client_cred=f"{client_id}:{client_secret}" #using an f-string
```

```python
#client ID and cleint secret must be in a base64 form (dont know why, but thats the way to do it)
client_cred_base64=base64.b64encode(client_cred.encode())
#client_cred_base64
```

```python
#prepare request input
# without the help of some tutorials online this would be impossible to get

auth_url = 'https://accounts.spotify.com/api/token'


data={"grant_type":"client_credentials"}

header={"Authorization": f"Basic {client_cred_base64.decode()}"} # the header has our access codes
#header
```

```python
# get an access token by sending a POST request
response=requests.post(auth_url, data=data, headers=header)
response_data=response.json()
response.status_code
## 200 means we are good
```

```python
#get the temporarily valid access token
access_token=response_data['access_token']
access_token
```

```python
#endpoint for our playlist( last number is playlistID one can copy from spotify by sharing a playlist)
#url="https://api.spotify.com/playlist/5pQVH0EJssqt4KDXPQVwv3"

url="https://api.spotify.com/v1/playlists/5pQVH0EJssqt4KDXPQVwv3/tracks" #
get directly track info

headers= {"Authorization": "Bearer {token}".format(token=access_token)} #authorization has changed from absic to bearer
headers
```

```python
res=requests.get(url,headers=headers)
res
#res.json()
```

```python
results=res.json()
results
```
